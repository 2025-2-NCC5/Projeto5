4.2 - Nessa etapa é feito o tratamento de dados vazio.

```python
import pandas as pd

if 'df_combinado' in locals():
    # Colunas selecionadas para processamento
    colunas_para_processamento = [
        'orderId', 'createdAt_x', 'customer', 'totalAmount',
        'paymentMethod', 'deliveryTime', 'city', 'gender'
    ]

    # Mantemos apenas as colunas existentes no DataFrame
    colunas_para_processamento = [c for c in colunas_para_processamento if c in df_combinado.columns]

    # Exibindo valores ausentes antes do tratamento
    print("Valores ausentes antes do tratamento:")
    display(df_combinado[colunas_para_processamento].isnull().sum())

    # Tratamento de valores ausentes
    for col in colunas_para_processamento:
        if df_combinado[col].dtype in ['float64', 'int64']:
            df_combinado[col].fillna(df_combinado[col].median(), inplace=True)
        else:
            mode_val = df_combinado[col].mode()
            if not mode_val.empty:
                df_combinado[col].fillna(mode_val[0], inplace=True)

    # Conversão de datas e criação de novas features
    if 'createdAt_x' in df_combinado.columns:
        df_combinado['createdAt_x'] = pd.to_datetime(df_combinado['createdAt_x'], errors='coerce')
        df_combinado['order_day_of_week'] = df_combinado['createdAt_x'].dt.dayofweek
        df_combinado['order_hour'] = df_combinado['createdAt_x'].dt.hour
        df_combinado['order_month'] = df_combinado['createdAt_x'].dt.month

    # Exibindo valores ausentes após limpeza
    print("\nValores ausentes após tratamento:")
    display(df_combinado[colunas_para_processamento].isnull().sum())

    # Mostrando primeiras linhas após tratamento
    print("\nPrimeiras linhas após limpeza e feature engineering:")
    display(df_combinado[colunas_para_processamento].head())

else:
    print("Erro: DataFrame combinado não encontrado. Impossível processar dados.")
```

<h1>Resumo</h1>

<p>Durante o carregamento inicial, os arquivos JSON apresentaram inconsistências e exigiram ajustes manuais, sendo possível combinar apenas os dados CSV de pedidos e clientes. A partir desses dados, identificamos várias variáveis potencialmente importantes para prever cancelamentos, incluindo informações do pedido, dados temporais e atributos dos clientes. Valores ausentes foram tratados com sucesso, como os 465 registros sem gênero, preenchidos com a moda.

A engenharia de atributos gerou novas informações relevantes: extraímos o dia da semana, a hora e o mês de criação dos pedidos, além de calcular a idade dos clientes com base na data de nascimento. A variável alvo is_cancelled foi definida a partir do status "CANCELED", contabilizando 312 pedidos cancelados entre 2005 registros. A codificação one-hot transformou colunas categóricas em atributos numéricos, aumentando a dimensionalidade para 3245 colunas.

O dataset foi dividido em treino (1508 amostras) e teste (497 amostras) usando estratificação para manter a proporção da variável alvo. Após remover colunas não numéricas, o modelo Gradient Boosting Classifier foi treinado com sucesso. A avaliação mostrou boa acurácia geral (0.839), mas dificuldade em prever cancelamentos, com precisão 0.021, recall 0.028, F1-Score 0.024 e ROC AUC 0.515, indicando que o modelo é eficaz em identificar pedidos não cancelados, mas ainda tem limitações para detectar cancelamentos.</p>
