
# Construção de um Modelo de Regressão Linear
**Disciplina:** Álgebra Linear  
**Tema:** Aplicação do Método dos Mínimos Quadrados  
**Base de Dados:** Estruturada na 1ª entrega (Desempenho dos Estudantes por Grupo)



## Objetivo

Aplicar a **Regressão Linear Simples** utilizando decomposição matricial (método dos mínimos quadrados), para modelar a relação entre:

- **Número total de estudantes (x)**  
- **Número de aprovados (y)**  

A partir desse modelo, estimar a quantidade prevista de aprovados para um novo valor de total de estudantes informado pelo usuário.


## 1 - Importação das Bibliotecas

````python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

## 2 - Definição dos Dados

````python

# Dados baseados na 1ª entrega
dados = {
    "Grupo": ["Frontend", "Backend", "Data Science", "Não Selecionado"],
    "Total": [210, 180, 160, 50],
    "Aprovados": [150, 120, 100, 20],
    "Reprovados": [60, 60, 60, 30]
}

df = pd.DataFrame(dados)
df
````


## 3 - Formulação Matemática

O modelo da **Regressão Linear Simples** é dado por:

\[ y = a + b \cdot x \]

onde:  
- \( y \): variável dependente (Aprovados)  
- \( x \): variável independente (Total de Estudantes)  
- \( a \): intercepto  
- \( b \): coeficiente angular (inclinação)

Usamos o método dos mínimos quadrados, resolvendo:

\[ \theta = (X^T X)^{-1} X^T y \]

onde \( X = [1, x] \).


## 4 - Implementação via Decomposição Matricial

````python

# Variáveis independentes e dependentes
x = df["Total"].values
y = df["Aprovados"].values

# Montagem da matriz X com coluna de 1's (intercepto)
X = np.vstack([np.ones(len(x)), x]).T

# Aplicação do método dos mínimos quadrados
theta = np.linalg.inv(X.T @ X) @ X.T @ y

# Coeficientes resultantes
a, b = theta
print(f"Intercepto (a): {a:.2f}")
print(f"Inclinação (b): {b:.2f}")
````

## 5 - Função de Previsão

````python

def prever_aprovados(total_estudantes):
    return a + b * total_estudantes

# Exemplo de uso
novo_total = 200
previsao = prever_aprovados(novo_total)
print(f"Para {novo_total} estudantes, a previsão de aprovados é: {previsao:.1f}")
````

## 6 - Visualização Gráfica

````python

plt.figure(figsize=(8,5))
plt.scatter(x, y, color='blue', label='Dados Reais')
plt.plot(x, a + b*x, color='red', label='Regressão Linear')

plt.title("Relação entre Total de Estudantes e Aprovados")
plt.xlabel("Total de Estudantes")
plt.ylabel("Número de Aprovados")
plt.legend()
plt.grid(True)
plt.show()
````


## 7 - Interpretação dos Resultados

- O **coeficiente angular (b)** mostra o quanto o número de aprovados cresce a cada novo estudante.  
- O **intercepto (a)** representa o valor teórico de aprovados caso não houvesse nenhum estudante (ajuste da reta).  
- A **reta de regressão** representa a melhor aproximação linear possível dos dados reais, calculada por meio de operações matriciais.



## 8 - Conclusão Analítica

- Foi possível modelar matematicamente a relação entre total de estudantes e aprovados.  
- O cálculo do vetor de parâmetros \((X^T X)^{-1} X^T y\) ilustra o uso direto da Álgebra Linear em regressões.  
- O modelo permite previsões quantitativas de aprovados em novos cenários.  
- A visualização gráfica confirma a correlação linear positiva entre as variáveis.
